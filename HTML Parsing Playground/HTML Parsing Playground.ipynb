https://info.cern.ch/hypertext/WWW/TheProject.html

```python
#load the required libraries 
from bs4 import BeautifulSoup as soup 
import requests
```

```python
#load data 
response=requests.get("https://info.cern.ch/hypertext/WWW/TheProject.html")
response.status_code
```

```python
#always check the type of response you are getting 
response.headers  #here content type : text/html
```

```python
html=response.text
#parse html data 
data=soup(html,'html.parser')
```

```python
#extract data 
#print heading 
#using isnpect tool highlight the part you want using select tool and use that 
#on clicking WWW heading trhough select tool it shows it is inside h1 so 
data.h1
```

```python
data.h1.string
```

```python
#to get the title 
data.title.string
```

```python
#to get complete text from this page 
data.get_text()
```

```python
#using print for making it more readable 
print(data.get_text())
```

```python
#extracting all hyperlinks from this web page 
for i in data.find_all('a'):
    print(i['href'])
```

```python
#let's try to extract links only from the specific section -> what's out there to getting a code and no links before that 
#in inspect tool we can say this part is actually a list under dl tag 
a=data.dl.find_all('a')
print(len(a))
for i in a:
    print(i['href'])
    
```

```python
#to print only links in heading of these parts note point title are in dt and others in dd 
a=data.dl.find_all('dt')
len(a)
```
