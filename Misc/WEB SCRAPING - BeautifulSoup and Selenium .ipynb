# WEB SCRAPING 

Web scraping is a technique for extracting information from the internet automatically using our script that simulates human web surfing.

Web surfing help us extract large volumes of data from different websites.

When data is present on a local system we can use CSV,TxT format and use numpy,pandas,SQL to analyze them but when data is present online only then one method is using API but not all websites provide api to fetch data. So one choice is to manually visit,locate the desired data and extract the data in some file or DB, but it will fail when we need to extract large amount of data sat 100's of wesbite's data so another way is **Automated** manner => Web scraping. 

Eg: For my analysis I want to list down each and every educational institute in delhi, it can be school,college,coaching institute or a coaching centre , all this data can be found on just dial.com, we can visit to extract all data but manual extraction for all of these is still hectic so we will write a code to go and scrape justdial.com to tell all the desired data from this website and then save in some file for further analysis.

SCRAPING RULES:

1)  Check a Website's Terms and Condition before you scrape it-> not all websites allow scraping.
2)  Do not spam the website by making a lot of requests to a specific web page (don't slow down their server)
3)  Update your code time to time as websites generally change their page structure time to time.

LIBRARIES USES (in python):

1)  BeautifulSoup -> can parse html and XML documents 
2)  Selenium -> with which we can initiate different browser sessions -> automate browser activities 
3)  Scrapy 

Process:

1)  Find the url that you want to scrape 
2)  Send an HTTP request to that URL and get the HTML as response 
3)  Parse the HTML content 
4)  Inspect the web page and find data that we want to extract 
5)  Extract required data and store it's data in the required format 

Note: Four major types of data are returned by server when called via get request : HTML,CSS,Javascript and Media content.
HTML will be majorly useful for extracting data since it has the main content (css is for styling, JS is for interaction and media content has images/videos and all), please revise HTML before proceeding

Links of all the pages 

Earlier we extracted data from first page only, this website has data of 1000 books -> available on 50 different pages. To get info about all the books we need to load data from all the pages 

The browse url should change for every get request, if you inspect the next button you will see the relative url to next pages each time you click on next , 
for page 2 it has class="next" href= "catalogue/page-2.html"  and after that just /page3.html , /page4.html. 
so 1st relative url need to have catalogue/ too in the link, other don't -> we can use an if else situation or we can once see what loads for books.toscrape.com/catalogue/page1.html rather than just for books.toscrape.com, page loading is same so better to use former one as base url

Note: if you think rather than extracting url from next we can just do base url+ page[i] i randing from 2 to 50 then this can only work here since page naming is systematic but what if page 2 has url named xyz and 3 has abc, so in that case better to make base url + relative url extracted from next button.

Also here we know limit for loop is 50 since total pages are 50 but in case we don't know then we can use while response.status_code= success that is 200

```python
import requests
from bs4 import BeautifulSoup as soup

#making a list of all urls
all_urls=["https://books.toscrape.com/catalogue/page-1.html"]
current_page='https://books.toscrape.com/catalogue/page-1.html'
base_url='https://books.toscrape.com/catalogue/'
response=requests.get(current_page)
while response.status_code==200:
    data=soup(response.text,'html.parser')
    next_page=data.find(class_='next')
    next_page_url=base_url+next_page.a['href']
    print(next_page_url)
    all_urls.append(next_page_url)
    current_page=next_page_url
    response=requests.get(current_page)
    
```

```python
#error at the end due to next_page=data.find(class_='next') since there is no class named next at the last page so to fix that:
all_urls=["https://books.toscrape.com/catalogue/page-1.html"]
current_page='https://books.toscrape.com/catalogue/page-1.html'
base_url='https://books.toscrape.com/catalogue/'
response=requests.get(current_page)
while response.status_code==200:
    data=soup(response.text,'html.parser')
    next_page=data.find(class_='next')
    if next_page is None:
        break
    next_page_url=base_url+next_page.a['href']
    print(next_page_url)
    all_urls.append(next_page_url)
    current_page=next_page_url
    response=requests.get(current_page)
```

From this website :
Find and print the src of the img tag which occupies the maximum area on the page.
Note :
Ignore images which doesn't have height or width attributes
Output Format :
src_of_image

link:https://en.wikipedia.org/wiki/Artificial_intelligence

```python

import requests 
from bs4 import BeautifulSoup as soup 
import re 

response=requests.get("https://en.wikipedia.org/wiki/Artificial_intelligence")
data=soup(response.text,"html.parser")
#first we can load all img tags 
maxarea=0
maxsrc=""
image_list=data.find_all('img')
for i in range(len(image_list)):
    if "height" in  image_list[i].attrs and "width" in  image_list[i].attrs :  #only if attrs has height and width 
        #to find area height Xwidth 
        area=int(image_list[i]["height"]) * int(image_list[i]["width"])
        if area>maxarea:
            maxarea=area
            maxsrc=image_list[i]["src"]
print(maxsrc)
```

# SELENIUM

There is one major issue with BeautifulSoup and that is we can only extract *static* content, content which doesn't require any kind of interaction, clicking on button,hyperlink,filling some form, entering some data into some text field or navigating through different webpages.

So limitation of BeautifulSoup:
1)  This process is suitable for static content only 
2)  Sometimes the data we want ti extract is hidden behind Javascript objects, objects that need to be clicked on to reveal the hidden data.

eg: from a certain website if i want to extract user comment then i need to first click on the link *show comments* -> won't display comment unless we click on it, BeautifulSoup can't help so first click to display content and then extract data ->selenium can do that
